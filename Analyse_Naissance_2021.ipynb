## Traitement des données abérrantes

```python
import pandas as pd
import numpy as np

def detect_abberation(df, n_std=3):
    mean = np.mean(df)
    std = np.std(df)

    lower_limit = mean - (n_std * std)
    upper_limit = mean + (n_std * std)

    return df[(df < lower_limit) | (df > upper_limit)]

aber_agemere = detect_abberation(df['AGEMERE'])
aber_agepere = detect_abberation(df['AGEPERE'])

aber_agemere, aber_agepere
```

Il n'y a aucune valeur abberante dans nos deux varibles père et mère s'il y'en avait, les valeurs aberrantes seraient traitées de différentes manières, telles que leur suppression ou leur remplacement par une valeur médiane ou moyenne.

### Tableau d’effectifs (à la manière du khi deux) : sexe bébé + Age de la mère pour voir si l'âge de la mère influ sur le sexe du de l'enfant 

```python
import pandas as pd
from scipy.stats import chi2_contingency

bins = [16, 25, 34, 46]
labels = ['17-25', '26-34', '35-46']
df.loc[:, 'CategorieAgeMere'] = pd.cut(df['AGEMERE'], bins=bins, labels=labels, right=True)
min_count = df['CategorieAgeMere'].value_counts().min()

equilibre_df = pd.DataFrame()
for category in labels:
    echantill_df = df[df['CategorieAgeMere'] == category].sample(n=min_count, random_state=1)
    equilibre_df = pd.concat([equilibre_df, echantill_df])

tableau_effectifs_equilibre = pd.crosstab(equilibre_df['SEXE'], equilibre_df['CategorieAgeMere'])
chi2_equilibre, p_equilibre, dof_equilibre, equilibre_attendu = chi2_contingency(tableau_effectifs_equilibre)
```

```python
tableau_effectifs_equilibre
```

```python
print("Résultat du chi2 :", chi2_equilibre)
print("Le degre de liberté:", dof_equilibre)
print("Valeur-p :", p_equilibre)
```

La valeur p est de 0.8890 qui est supérieure à 0.05, cela indique qu'il n'y a pas de preuve statistique significative pour rejeter l'hypothèse nulle.
L'hypothèse nulle est que les deux variables catégorielles sont indépendantes l'une de l'autre. Donc la distribution du sexe des bébés est indépendante de l'âge des mères dans l'échantillon étudié, après avoir équilibré les effectifs par tranche d'âge.

```python
import pandas as pd
from scipy.stats import chi2_contingency

bins = [16, 25, 34, 46]
labels = ['17-25', '26-34', '35-46']
df.loc[:, 'CategorieAgePere'] = pd.cut(df['AGEPERE'], bins=bins, labels=labels, right=True)
min_count = df['CategorieAgePere'].value_counts().min()

equilibre_df = pd.DataFrame()
for category in labels:
    echantill_df = df[df['CategorieAgePere'] == category].sample(n=min_count, random_state=1)
    equilibre_df = pd.concat([equilibre_df, echantill_df])

tableau_effectifs_equilibre = pd.crosstab(equilibre_df['SEXE'], equilibre_df['CategorieAgePere'])
chi2_equilibre, p_equilibre, dof_equilibre, equilibre_attendu = chi2_contingency(tableau_effectifs_equilibre)
```

```python
tableau_effectifs_equilibre
```

```python
print("Résultat du chi2 :", chi2_equilibre)
print("Le degre de liberté:", dof_equilibre)
print("Valeur-p :", p_equilibre)
```

Il n'existe pas de lien statistiquement significatif entre l'âge du père au moment de la conception et le sexe de l'enfant à la naissance.

### Faire un test de normalité pour chaque histo mere et pere

```python
from scipy.stats import shapiro
stat_mere, p_mere = shapiro(df['AGEMERE'].sample(5000, random_state=1))  # Échantillonnage à cause de la limite de taille
stat_pere, p_pere = shapiro(df['AGEPERE'].sample(5000, random_state=1))  # Échantillonnage à cause de la limite de taille

print(f"Statistique de test pour l'âge des pères: {stat_pere}, P-value: {p_pere}")
print(f"Statistique de test pour l'âge des mères: {stat_mere}, P-value: {p_mere}")
```

Les p-values sont inférieures à 0.05 alors les données suivent une distribution normale
